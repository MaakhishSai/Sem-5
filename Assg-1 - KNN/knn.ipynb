```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# add an r as we are reading it as raw string or else add '\' before every partition of the path
df = pd.read_csv(r"D:\Sem - 5\ML- LAB\Assg-1 - KNN\fruit_dataset.csv")

# print(df.head())

x = df.iloc[ : , 2: ] # all input features
y = df.iloc[ : , 0:1] # class labels

x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 1234, test_size = 0.25, shuffle = True)
```

```python
def knn(x_train, x_test_sample, k, y_train):
    x_test_sample = np.array(x_test_sample)
    distances = []
    for i in range(len(x_train)):
        x_train_sample = np.array(x_train.iloc[i])
        label = y_train.iloc[i, 0]
        dist = 0
        for j in range(len(x_train_sample)):
            dist += (x_train_sample[j] - x_test_sample[j])**2
        dist = dist**0.5
        distances.append((dist, label))
    
    distances.sort(key = lambda x:x[0])

    k_nearest_labels = [label for _,label in distances[ : k]]
     
    return k_nearest_labels
```

```python
def fun(k_nearest_labels, x_test_sample, x_train, y_train, k):
    # most common label
    k_nearest_labels = knn(x_train, x_test_sample, k, y_train)
    label_counts = {}
    for label in k_nearest_labels:
        if label in label_counts:
            label_counts[label] += 1
        else:
            label_counts[label] = 1
    
    print(label_counts)
    
    most_common_label = None
    max_count = 0
    for label, count in label_counts.items():
        if count >= max_count:
            max_count = count
            most_common_label = label
    return most_common_label
```

```python
k = 5
tc = len(x_test)
tcor = 0
for ind in range(len(x_test)):
    x_test_sample = np.array(x_test.iloc[ind])
    ("X_test_sample : ")
    print(x_test_sample)
    knns = knn(x_train, x_test_sample, k, y_train)
    predicted_label = fun(knns, x_test_sample, x_train, y_train, k)
    actual_label = y_test.iloc[ind, 0]
    print("Actual Label : ", actual_label)
    print("Predicted Label : ", predicted_label)
    if actual_label == predicted_label:
        tcor += 1
    print('-----------------------------------------------------')

acc = tcor / tc

print(f"Accuracy of the model is : {acc*100 : .4f}")
```
